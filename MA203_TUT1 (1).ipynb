**QUESTION 1**

Fibonacci Sequence Generation


```python
def fibonacci(x):
  if x==1:
    return 0
  if x==2 :
    return 1
  else:
    return fibonacci(x-1)+fibonacci(x-2)

N=int(input())
for i in range(1,N+1):
  y=fibonacci(i)
  print(y,end=" ")
```

**QUESTION 2**

Finding smallest number

```python
sets=list(map(int,input().split()))
min=sets[0]
for i in range(len(sets)):
  if sets[i]<min:
    min=sets[i]
print(min)
```

**QUESTION 3**

Finding factors

```python
sets=list(map(int,input().split()))
for i in sets:
  print("factor of",i,":")
  for j in range(1,i+1):
    if (i%j==0):
      print(j,end=" ")
  print("\n")
```
